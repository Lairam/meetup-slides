# Image processing

```python
import numpy as np
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
```

```python
img = Image.open('assets/urban.jpg').convert('L')
plt.figure()
plt.imshow(img, cmap='gray')

img_array = np.asarray(img)

def conv2d(img_array, kernel):
    conv2d_img = convolve2d(img, kernel)
    figs, ax = plt.subplots(1, 2, figsize=(15, 5))
    ax[0].imshow(img_array, cmap='gray')
    ax[0].set_title('original')
    ax[1].imshow(conv2d_img, cmap='gray')
    ax[1].set_title('convolved')
```

```python
kernel = np.array([
    [ 1,  1,  1],
    [ 1,  1,  1],
    [ 1,  1,  1],
])

conv2d(img_array, kernel)
```

```python
kernel = np.array([
    [ 1,  1,  1],
    [ 1,  1,  1],
    [ 1,  1,  1],
])

mu = 0
sigma = 50
noisy_img_aray = img_array + np.random.normal(mu, sigma, img_array.shape)
conv2d(noisy_img_aray, kernel)
```

```python
kernel = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1],
])

conv2d(img_array, kernel)
```

```python
kernel = np.array([
    [-1,  0,  1],
    [-2,  0,  2],
    [-1,  0,  1],
])

conv2d(img_array, kernel)
```

```python
kernel = np.array([
    [ 0,  1,  0],
    [ 1, -4,  1],
    [ 0,  1,  0],
])

conv2d(img_array, kernel)
```
